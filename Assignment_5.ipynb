#Question No.1 List out the features of HTML 5.
1. New Semantic Elements-<br>
  --> HTML5 introduced more descriptive tags that define the structure of a webpage more clearly:<br>
  <ul>
   <li>header, footer</li>
   <li>article, section</li>
   <li>nav, aside, main</li>
   </ul>

2. Multimedia Support-<br>
  --> Built-in support for audio and video without needing external plugins like Flash:<br>
  <ul>
  <li>Audio</li>
  <li>Video</li>
  </ul>

3. Graphics and Visuals-<br>
  <ul>
  <li>canvas: for drawing graphics on the fly (e.g., games, charts)</li>
  <li>svg: for scalable vector graphics.</li>
  </ul>

4. Responsive Design Support-  
  <ul>
  <li> Improved support for mobile devices.</li>
  <li> Media queries via CSS3 (though not HTML5 directly, they work hand-in-hand)</li>
  </ul>

5. Storage Options-<br>
 <ul>
  <li>Local Storage: localStorage and sessionStorage allow storing data on the client-side</li>
  <li>Offline Web Applications: via the Application Cache (deprecated in favor of Service Workers)</li>
  </ul>

#Question No.3  Define accessibility in the context of web development. Discuss why it's essential to create accessible websites and how it benefits different user group.
#Answer-
1. Accessibility in Web Development-<br>
<ul>
 <li>Accessibility in web development refers to designing and building websites and web applications so that all users, including those with disabilities, can perceive, understand, navigate, and interact with them effectively. This includes people with:</li>

      <ul>
        <li>Visual impairments (e.g., blindness, color blindness)</li>
        <li>Hearing impairments</li>
        <li>Motor disabilities</li>
        <li>Cognitive or learning disabilities</li>
        <li>Temporary impairments (e.g., a broken arm)</li>
        <li>Situational limitations (e.g., bright sunlight or slow internet)</li>
      </ul>
</ul>  
2.Why Accessibility Is Essential-<br>
(a)Equal Access for Everyone-<br>
   -->It ensures that all users, regardless of their abilities, can access and use online content.

  (b) Legal Compliance-<br>
   -->Many countries have laws and regulations (like the ADA in the U.S. or WCAG guidelines globally) that require digital accessibility.

  (c) Improved User Experience-<br>
   -->Accessible websites are often more usable for everyone, with better structure, navigation, and clarity.

  (d) Better SEO-<br>
   -->Accessible features like alternative text for images and clear headings also help search engines understand your content better.

  (e) Wider Audience Reach--<br>
   -->By designing inclusively, you reach a larger and more diverse audience, including elderly users or people with temporary challenges.<br>

#Question No.4 List any 3 ways which help us in improving the accessibility of HTML.
#Answer-
1. Use Semantic HTML Elements:<br>
   --> Semantic tags (like "header","nav", "main", "section", "article", and "footer") help screen readers and other assistive technologies understand the structure and meaning of the content.
2. Provide Alternative Text for Images (alt attribute)<br>
  --> Use the "alt" attribute in "img" tags to describe images. This helps screen readers convey visual content to visually impaired users.
3. Ensure Proper Labeling of Form Elements<br>
   -->Use "label" tags associated with form inputs so screen readers can identify the purpose of each input field.
